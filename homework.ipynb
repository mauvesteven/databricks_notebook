<b> Load the CSV file with datatime parser </b>

```
dt_parser = lambda x: pd.to_datetime(x, format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

df = pd.read_csv('Fire_Department_Calls_for_Service_sampled_1_percent.csv',
                 sep         = ",", 
                 skiprows    = 1, 
                 names       = col_names, 
                 parse_dates = dt_col_names, 
                 date_parser = dt_parser)

# parse the 'Call Date' column as well
df['Call_Date_Dt'] = pd.to_datetime(df['Call Date'], format='%m/%d/%Y')

# deploy in seconds
df['Delay'] = (df['On Scene DtTm']-df['Received DtTm'])/ np.timedelta64(1, 's')

df.head()
```

# Question 3: What’s the average delay (defined as the difference between a call’s receive time and the on scene time) for all calls in 2019? What’s the 99%-tile delay?
Answer: average delay: 586 seconds,  99%-tile delay 2294 seconds (Please find the SQL query below)

Please note that multiple units might be dispatched for the same call and each unit have its own on scene time, the statistics are computed based on <b> all the distinct dispatched units instead of all the calls </b>.

```
%sql
select count(*) from sparkView
  where year(Call_Date_Dt) == 2019
```

```
%sql
select count(*) from sparkView
  where year(Call_Date_Dt) == 2019 and isnull(`On Scene DtTm`)
```

```
%sql
select count(*) from sparkView
  where year(Call_Date_Dt) == 2019 and isnull(`Received DtTm`)
```

```
%sql
select min(Delay), percentile(Delay, 0.01), mean(Delay), percentile(Delay, 0.99), max(Delay), count(*)
  from sparkView 
  where year(Call_Date_Dt) == 2019 and !isnull(`On Scene DtTm`)
```

Note that there are some calls that has ZERO delay which could be a data quality issue? For the sake of time, I won't further investigate into it

```
%sql
select *
  from sparkView 
  where year(Call_Date_Dt) == 2019 and !isnull(`On Scene DtTm`) and delay == 0
```

# Question 4: Are the number of calls trending up, or trending down, on a year-over-year basis?
Answer: Overall speaking, number of calls are trending up, except a sudden increase in 2009 and then decrease in 2010. (Pleaes find the corresponding visualization below)

Note that this does not taking year 2010 and 2019 into account because we don't have data for all the months for those two years.

```
%sql
create table table_distinct_calls as
  select distinct `Call Number` as Call_Number, Call_Date_Dt, year(Call_Date_Dt) as year
  from sparkView
```

```
%sql
select year, count(*) from table_distinct_calls
  group by year
  order by year
```
